```python
import requests

def asset_unit_market_value(client,asset_unit_id,start_date,end_date):
    holding_url = '{}/api/rqams/v1/asset_units/{}/market_value?'.format('http://rqams.ricequant.pro',asset_unit_id)
    params = dict(start_date = start_date,end_date = end_date)
    cookies = dict(sid=client.sid)
    r = requests.get(holding_url,params=params,cookies=cookies)
    return r.json()
```
